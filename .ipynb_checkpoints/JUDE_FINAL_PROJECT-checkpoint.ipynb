### Exploratory Data Analysis

```python
def generate_bar_plot(x,y, xlabel, ylabel):
    plt.bar(x,y)
    plt.title(f'plot of {ylabel}against{xlabel}')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show
```

```python
Df_exited = Df.groupby('Geography')[['Exited']].sum()
Df_exited = Df_exited.sort_values(by='Exited', ascending=False)
Df_exited
```

```python
generate_bar_plot(Df_exited.index, Df_exited['Exited'], 'Geography', 'Exited')
```

```python
Result = Df.groupby('Gender')[['Has Credit Card']].count()
```

```python
Result
```

```python
generate_bar_plot(Result.index, Result['Has Credit Card'], 'Gender', 'Has Credit Card')
```

```python
Customer_dist = Df.groupby('Gender')[['CustomerId']].size().reset_index(name ='Customer_Count')    
Customer_dist
```

```python
def plot_pie_chart(df, labels_column, count_column, title):

    # Get the labels and values
    labels = df[labels_column]
    sizes = df[count_column]

    # Create a pie chart
    plt.figure(figsize=(20,10))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightcoral'])
    plt.title(title)
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

    # Show the plot
    plt.show()
```

```python
plot_pie_chart(Customer_dist, labels_column='Gender', count_column='Customer_Count', title='Customer Distribution by Gender')
```
